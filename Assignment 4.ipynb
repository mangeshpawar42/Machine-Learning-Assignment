```python
import nltk
def learnRETagger(simpleSentence):
  customPatterns = [
  (r'.*ing$', 'ADJECTIVE'),
  (r'.*ly$', 'ADVERB'),
  (r'.*ion$', 'NOUN'),
  (r'(.*ate|.*en|is)$', 'VERB'),
  (r'^an$', 'INDEFINITE-ARTICLE'),
  (r'^(with|on|at|from)$', 'PREPOSITION'),
  (r'^\-?[0-9]+(\.[0-9]+)$', 'NUMBER'),
  (r'^(i|his|her)$', 'PRONOUN'),
  (r'^(a|an|the)$', 'ARTICLE'),
  (r'.*$', None),
  ]
  tagger = nltk.RegexpTagger(customPatterns)
  wordsInSentence = nltk.word_tokenize(simpleSentence)
  posEnabledTags = tagger.tag(wordsInSentence)
  print(posEnabledTags)
```

```python
#sentence = input("Enter your sentence: ")
sentence = "My name is Jayendra and I am pursuing BTech from the VIIT"
sentence1 = sentence.lower()
learnRETagger(sentence1)
```
