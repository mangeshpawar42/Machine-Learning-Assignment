```python
import nltk
nltk.download('punkt')
```

```python
def RDParserExample(grammar, textlist):
    parser = nltk.parse.RecursiveDescentParser(grammar)
    for text in textlist:
        sentence = nltk.word_tokenize(text)
        for tree in parser.parse(sentence):
            print(tree)
            tree.draw()
```

```python
def SRParserExample(grammar, textlist):
    parser = nltk.parse.ShiftReduceParser(grammar)
    for text in textlist:
        sentence = nltk.word_tokenize(text)
        for tree in parser.parse(sentence):
            print(tree)
            tree.draw()
```

```python
grammar1 = nltk.CFG.fromstring("""
S -> NP-SBJ VP
NP-SBJ -> NP PP 
VP -> VBD
VBD -> 'passed'
NP -> QP | DT NNS
PP -> IN NP
QP -> IN JJS CD NNS
IN -> 'of' | 'at'
DT -> 'the' | 'The'
JJS -> 'least'
CD -> 'nine'
NNS -> 'tenths' | 'students'
""")

text1 = [    "at least nine tenths of the students passed",
]



grammar2 = nltk.CFG.fromstring("""
S -> NP VP
NP -> Det N | Det N PP
VP -> V NP
PP -> P NP
Det -> 'the' | 'a'
N -> 'dog' | 'man' | 'park'
V -> 'saw'
P -> 'in'
""")

text2 = ["the dog saw a man in the park"]

grammar3 = nltk.CFG.fromstring("""
S -> NP VP
NP -> DT NNS | NNP
VP -> VBD ADVP VP
DT -> 'the'
NNS -> 'cells'
VBD -> 'were'
ADVP -> RB
VP -> VBN PP
RB -> 'previously'
VBN -> 'infected'
PP -> IN NP
IN -> 'with'
NNP -> 'HIV-1'
""")

text3 = ["the cells were previously infected with HIV-1"]

grammar4 = nltk.CFG.fromstring("""
S -> NP VP
NP -> PRON N | Det N
VP -> V PRON PP
PP -> PREP NP
PRON -> 'My' | 'me'
N -> 'shoes' | 'soil'
V -> 'separate'
PREP -> 'from'
Det -> 'the'
""")

text4 = ["My shoes separate me from the soil"]

grammar5 = nltk.CFG.fromstring("""
S -> NP VBD IN NP
NP -> DT JJ JJ NN | DT NN
VBD -> 'barked'
IN -> 'at'
DT -> 'the' 
JJ -> 'little' | 'yellow' 
NN -> 'dog' | 'cat'
""")

text5 = ["the little yellow dog barked at the cat"]
```

```python
RDParserExample(grammar1, text1)
SRParserExample(grammar1, text1)
```

```python
RDParserExample(grammar2, text2)
SRParserExample(grammar2, text2)
```

```python
RDParserExample(grammar3, text3)
SRParserExample(grammar3, text3)
```

```python
RDParserExample(grammar4, text4)
SRParserExample(grammar4, text4)
```

```python
RDParserExample(grammar5, text5)
SRParserExample(grammar5, text5)
```
